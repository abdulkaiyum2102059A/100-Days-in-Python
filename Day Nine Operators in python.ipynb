# Operators

Operators are used to perform operations on variables and values.Python has the following operators:

-Arithmetic operators 
-Comparison operators
-Logical operators
-Bitwise operators
-Assignment operators
-Identify operators
-Membership operators


```python
# Arithmetic Operators
x = 5
y = 2

print(x+y)
```

```python
print(x-y)
```

```python
print(x*y)
```

```python
print(x/y)
```

```python
print(x%y)
```

```python
print(x ** y)
```

```python
print(x // y)
```

```python
# Comparison Operators
print(x<y)
```

```python
print(x>y)
```

```python
print(x>=y)
```

```python
print(x<=y)
```

```python
print(x==y)
```

```python
print(x != y)
```

```python
#Logical Operators

x = True
y = False 

print (x or y)
```

```python
print (x and y)
```

```python
print (not y)
```

```python
# Bitwise 

x = 2 
y = 3

print ( x & y)
```

```python
print (x | y)
```

```python
print(~x)
```

```python
#Assignment Operators
a = 3
print(a)
```

```python
a +=3
#a = a + 3
print(a)
```

```python
a -= 3
a *= 3
a &= 3
```

```python
a = 3
b = 3
print (a is b)
```

```python
a = "Hello"
b = "Hello"
print (a is b)
```

```python
a = "Hello - World"
b = "Hello - World"
print (a is b)
```

```python
a = [1,2,3]
b = [1,2,3]
print (a is b)
```
