#Indentation

In Python, indentation is not just for readability — it defines the structure of the code. Python uses indentation to group statements into blocks (like loops, functions, and conditions).

Without correct indentation, your code will raise an IndentationError or behave incorrectly.

Rules of Python Indentation (Short)

Use 4 spaces per indentation level (recommended standard).

Be consistent (don’t mix spaces and tabs).

All statements in the same block must have the same indentation.

Indentation is required after:

if, elif, else

for, while

def

class

try, except, finally

Improper indentation causes an IndentationError.

```python
name = "abdul kaiyum"
if name == "abdul kaiyum":
    print("He is a student")
    print("His father name is Abu Taiyub")
    if 5 == 5:
        print('he is a student')
else:
    print('Jahidul Islam')
```
